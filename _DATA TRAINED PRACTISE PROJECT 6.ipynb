# Finding multicollinearity we apply vif

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
dfx.shape
```

```python
dfx.shape[1]
```

```python
def vif_calc():
    vif=pd.DataFrame() 
    vif['VIF Factor']=[variance_inflation_factor(dfx.values,i)for i in range(dfx.shape[1])]
    vif['FEATURES']=dfx.columns
    print(vif)
```

```python
vif_calc()
```

# From the above study it is understood , we should remove Seat_No. as it variance_inflation_factor is too high and does not suit the datasets values.

# Training process

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

# Linear model selected as output variable contains more of float data

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

# Getting the right random_state

```python
maxAcc=0
maxRS=0
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(dfx,y,test_size=0.20,random_state=i)
    lr.fit(x_train,y_train)
    predlr=lr.predict(x_test)
    score=r2_score(y_test,predlr)
    print('R2_score of LinearRegression ' ,score, 'at random_state ' ,i)
    if score > maxAcc:
        maxAcc=score
        maxRS=i
    print('Mac acc score of lr ' ,maxAcc*100, 'at random_state ' ,maxRS)    
```

```python
x_train,x_test,y_train,y_test=train_test_split(dfx,y,test_size=0.20,random_state=246)
lr.fit(x_train,y_train)
predlr=lr.predict(x_test)
score=r2_score(y_test,predlr)
print('R2_score of LinearRegression ' ,score*100)
print("error :")
print("Mean absolute error: ",mean_absolute_error(y_test,predlr))
print("Mean squared error: ",mean_squared_error(y_test,predlr))
print("Square root of mean squared error:", np.sqrt(mean_squared_error(y_test,predlr)))
```

# Regularisation for regression

```python
from sklearn.linear_model import Lasso,Ridge,ElasticNet
l=Lasso()
r=Ridge()
e=ElasticNet()
```

```python
l.fit(x_train,y_train)
r.fit(x_train,y_train)
e.fit(x_train,y_train)
```

```python
mod=[l,r,e]
for i in mod:
    alpha=[0.0001,0.001,0.01,0.1,1]
    for j in alpha:
        predi=i.predict(x_test)
        print('R2_score' , r2_score(y_test,predi)*100,'at model ' ,i , ' with alpha ' ,j)
        print("Mean absolute error: ",mean_absolute_error(y_test,predi))
        print("Mean squared error: ",mean_squared_error(y_test,predi))
        print("Square root of mean squared error:", np.sqrt(mean_squared_error(y_test,predi)))
```

# Testing with more models

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor,RandomForestRegressor,AdaBoostRegressor,ExtraTreesRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
knr=KNeighborsRegressor()
gbr=GradientBoostingRegressor()
rfr=RandomForestRegressor()
abr=AdaBoostRegressor()
etr=ExtraTreesRegressor()
dtr=DecisionTreeRegressor()
svr=SVR()
```

```python
knr.fit(x_train,y_train)
gbr.fit(x_train,y_train)
rfr.fit(x_train,y_train)
abr.fit(x_train,y_train)
etr.fit(x_train,y_train)
dtr.fit(x_train,y_train)
svr.fit(x_train,y_train)
```

```python
models=[knr,gbr,rfr,abr,etr,dtr,svr]
for i in models:
    predmodel=i.predict(x_test)
    print('R2_score of model : ' ,i, '=' ,(r2_score(y_test,predmodel))*100)
    print("Mean absolute error: ",mean_absolute_error(y_test,predmodel))
    print("Mean squared error: ",mean_squared_error(y_test,predmodel))
    print("Square root of mean squared error:", np.sqrt(mean_squared_error(y_test,predmodel)))
```

# Comparing  the most  suitable  models

```python
a=np.array(y_test)
l=np.array(lr.predict(x_test))
e=np.array(etr.predict(x_test))
g=np.array(gbr.predict(x_test))
r=np.array(rfr.predict(x_test))
df_trial=pd.DataFrame(data={'Actual_results':a,'Linear':l,'ExtraTrees':e,'GradientBoosting':g,'RandomForest':r},index=range(len(a)))
df_trial
```
